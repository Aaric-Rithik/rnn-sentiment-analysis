```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    "the glass of milk",
    "the glass of juice",
    "the cup of tea",
    "I am a good boy",
    "I am a good developer",
    "understand the meaning of words",
    "your videos are good"
]
```

```python
sent
```

```python
# Define the Vocubulary size

voc_size = 10000
```

```python
# One hot representation with every word
one_hot_repr = [one_hot(words,voc_size) for words in sent]
```

```python
one_hot_repr
```

```python
## Word Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import pad_sequences
```

```python
import numpy as np
```

```python
sent_length = 8
embeded_docs = pad_sequences(one_hot_repr,padding='post',maxlen=sent_length) # Zeros will be added to the end
# embeded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embeded_docs)
```

```python
# Feature representation
dim =10
```

```python
# create embedding layer

# model = Sequential()
# model.add(Embedding(voc_size,dim,sent_length))
# model.compile('adam','mse') # Krish naik, following is GPT one..
#---
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Input

model = Sequential()
model.add(Input(shape=(sent_length,)))
model.add(Embedding(voc_size, dim))

model.compile(optimizer='adam', loss='mse')

```

```python
model.summary()
```

```python
model.predict(embeded_docs)
```
